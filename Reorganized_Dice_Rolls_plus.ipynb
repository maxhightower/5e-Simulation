```python
def levelToProficiency(Player_Character):
  count = 0
  for i in Player_Character.Levels:
      if i != False:
        count = count + 1
      else:
        pass
  return np.floor(2 + (count - 1)/4)
```

```python
def CRToProficiency(Monster):
  print(type(Monster.CR))
  return np.floor(2 + (Monster.CR - 1)/4)
```
